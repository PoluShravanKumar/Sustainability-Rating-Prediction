# Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
```

```python
cont_transform = Pipeline([("Mean Imputer",SimpleImputer(strategy= "mean")),
         ("Standard Scaler",StandardScaler())])
cont_transform
```

```python
nom_transform = Pipeline([("Mode Imputer", SimpleImputer(strategy= "most_frequent")),
         ("one hot encoding",OneHotEncoder(drop = 'first') )])
nom_transform
```

```python
ordinal_cat_cols
for i in ordinal_cat_cols:
    print(i,df[i].unique())
```

```python
ordinal_mapping = [
    ('ClothingFrequency', ['Rarely', 'Sometimes', 'Often', 'Always']),
    ('CommunityInvolvement', ['Low','Moderate' , 'High']),
    ('LocalFoodFrequency', ['Rarely', 'Sometimes', 'Often', 'Always']),
    ('PhysicalActivities', ['Low','Moderate' , 'High']),
    ('UsingPlasticProducts', ['Never', 'Rarely', 'Sometimes', 'Often'])
]
```

```python
ord_transform = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinal', OrdinalEncoder(categories=[mapping[1] for mapping in ordinal_mapping]))
])
ord_transform
```

# Preprocessing

```python
from sklearn.compose import  ColumnTransformer
```

```python
Preprocess = ColumnTransformer([("Continuation Tranformation", cont_transform, numerical_cols),
                  ("Nominal Tranformation", nom_transform, nominal_cat_cols),
                  ("Ordinal Tranformation", ord_transform, ordinal_cat_cols)])

Preprocess
```
